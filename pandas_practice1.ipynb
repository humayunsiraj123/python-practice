```python
#survive
def result(x,y):
    if x=="female":
        if y<1:
            return 90
    else:
        return 0
df2["surviv"]=df2[["Sex","SibSp"]].apply(lambda x,y:result(*x,*y),axis=1  )
df2.surviv.value_counts()
```

```python
def mapping(x):
    if x=="male" :return 1
    elif x=="female":return 2
    elif x==np.nan :return 0
    
df2["mapping"]=df2.Sex.apply(mapping)
df2[(df2.mapping)!=(df2.gender)]
```

```python

def grade(p):
    if p>= 80: return "A+"
    elif p>=70: return "A" 
    elif p>=60: return "B"
    elif p>=50: return "C"
    else: return "Fail"

df2["classes"]=df2.Age.apply(grade)
df2.classes.value_counts()
```
