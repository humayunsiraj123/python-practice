```python
#task 1 student  list
n=int(input("enter the  no of students"))
student={}
for i in range(n):
    name=input("enter the name of students:")
    marks=int(input("enter the  marks out of 100 :"))
    student.update({name:marks})
print(student)    
lst=[]    
for i in  student:
    if student[i]<45:
        lst.append(i)
for i in lst:
    del student[i]
print(student)    
```
